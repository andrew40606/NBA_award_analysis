## Draw Correlation Heatmap

```python
corr = MVP_data.drop(['WS_x', 'TRB%', 'ORB%', 'DRB%', 'BLK%', 'AST%', 'STL%', 'Share'], axis = 1).corr()
sns.set(rc = {'figure.figsize':(15, 10)})
sns.heatmap(corr)
plt.title('Heatmap of Player Stats')
plt.show()
```

## Comparison Between MVP Candidates and Other Players

### PER Comparison

```python
sns.set(rc = {'figure.figsize':(8, 6)})
sns.kdeplot(all_adv.PER, label = 'All Players')
sns.kdeplot(MVP_data.PER, label = 'MVP candidates')
plt.title('PER Distribution (MVP vs league)')
plt.legend()
plt.show()
```

### WS Comparison

```python
sns.kdeplot(all_adv['WS/48'], label = 'All Players')
sns.kdeplot(MVP_data['WS/48_x'], label = 'MVP candidates')
plt.title('WS Distribution (MVP vs league)')
plt.legend()
plt.show()
```

### VORP Comparison

```python
sns.set(rc = {'figure.figsize':(8, 6)})
sns.kdeplot(all_adv.VORP, label = 'All Players')
sns.kdeplot(MVP_data.VORP, label = 'MVP candidates')
plt.title('VORP Distribution (MVP vs league)')
plt.legend()
plt.show()
```

### Offensive v.s. Defensive (MVP Candidate v.s. Other Players

```python
sns.set(rc = {'figure.figsize':(12, 8)})
sns.scatterplot(x = all_adv.DBPM, y = all_adv.OBPM, label = 'all players', color = 'lightgreen', alpha = 0.5)
sns.scatterplot(x = MVP_data.DBPM, y = MVP_data.OBPM, label = 'MVP candidates', color = 'red', size = MVP_data.Share)
plt.title('OBPM-DBPM (MVP vs league)')
plt.legend()
plt.show()
```
